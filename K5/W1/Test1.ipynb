```python
import pandas as pd
milk = pd.read_csv('monthly-milk-production.csv',';', index_col=['month'], parse_dates=['month'], dayfirst=True)
```

```python
milk.plot()
```

```python
import statsmodels.api as sm
sm.tsa.stattools.adfuller(milk.milk)
```

```python
milk['daily'] = milk['milk'] / milk.index.days_in_month
milk.daily.plot()
```

```python
milk.daily.sum()
```

```python
milk.daily_diff1 = milk.daily - milk.daily.shift(1)
milk.daily_diff12 = milk.daily - milk.daily.shift(12)
milk.daily_diff12_1 = milk.daily_diff12 - milk.daily_diff12.shift(1)
print 'Start: ', sm.tsa.stattools.adfuller(milk.daily)[1]
print 'Diff: ', sm.tsa.stattools.adfuller(milk.daily_diff1.dropna())[1]
print 'Season: ', sm.tsa.stattools.adfuller(milk.daily_diff12.dropna())[1]
print 'Season + diff: ', sm.tsa.stattools.adfuller(milk.daily_diff12_1.dropna())[1]
```

```python
milk.daily_diff12_1.plot()
```

```python
import statsmodels.api as sm
sm.graphics.tsa.plot_acf(milk.daily_diff12_1.dropna().values.squeeze(), lags=50)
```

```python
sm.graphics.tsa.plot_pacf(milk.daily_diff12_1.dropna().values.squeeze(), lags=50)
```
